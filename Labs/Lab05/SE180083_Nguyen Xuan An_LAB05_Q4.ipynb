# Import

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
from io import StringIO
import numpy as np

url = 'https://en.wikipedia.org/wiki/Demographics_of_Russia'
response = requests.get(url)
html_content = response.text

soup = BeautifulSoup(response.text, 'html.parser')
tables = pd.read_html(response.text)
```

# EDA

```python
df_beforeww2 = tables[10]
df_afterww2 = tables[11]
```

```python
df_beforeww2.head()
```

```python
df_afterww2.head()
```

```python
df_afterww2.columns
```

```python
df_afterww2 = df_afterww2.drop(columns=['Abortions (including miscarriages) reported', 'Life Expectancy (total)', 
'Urban fertility', 'Rural fertility', 'Crude migration change (per 1,000)'], axis=1)
```

```python
df_beforeww2 = df_beforeww2.rename(columns={'Unnamed: 0': "Year", 
                                "Average population[35]": "Average population"})
```

```python
df_afterww2 = df_afterww2.rename(columns={'Unnamed: 0': "Year",
    "Total average population (January 1 1993 onwards)" : "Average population",
    "Total fertility rates[fn 2]": "Total fertility rates"
    })
```

```python
df_beforeww2.info()
```

```python
df_afterww2.head()
```

```python
df_afterww2.info()
```

```python
df_afterww2['Year'] = df_afterww2['Year'].replace("2014[a]", "2014")
df_afterww2['Average population'] = df_afterww2['Average population'].replace("148,561,694 [46]", "148561694")
```

```python
df_afterww2 = df_afterww2.astype({'Year': np.int64, "Average population": np.int64})
```

```python
df_pop = pd.concat([df_beforeww2, df_afterww2],axis=0, join="inner")
df_pop
```

```python
df_pop = df_pop.replace(np.nan, 0)
df_pop
```

# Visualize

```python
import dash
from dash import dcc, html
import plotly.express as px
import plotly.graph_objects as go
from dash.dependencies import Input, Output
```

```python
fig_population_beforeww2 = px.line(df_beforeww2, x='Year', y='Average population',
                        line_shape='linear',
                        title='Population of Russia before World War 2')

fig_population_afterww2 = px.line(df_afterww2, x='Year', y='Average population',
                        line_shape='linear',
                        title='Population of Russia after World War 2')

fig_population_over_time = go.Figure()
fig_population_over_time.add_trace(go.Scatter(x=df_pop['Year'], y=df_pop['Average population'],
                                            mode='lines+markers', name='Population'))
fig_population_over_time.update_layout(
    title='Population of Russia Over Time',
    xaxis={'title': 'Year'},
    yaxis={'title': 'Population (millions)'},
    hovermode='closest'
)

total_population_beforeww2 = df_beforeww2['Average population'].sum()
total_population_afterww2 = df_afterww2['Average population'].sum()

fig_pie_comparison = px.pie(names=['Before WW2', 'After WW2'],
                            values=[total_population_beforeww2, total_population_afterww2],
                            title='Pie chart of the population before & after World War 2',
                            hole=0.3)

correlation_matrix = df_pop.corr()
fig_correlation_heatmap = px.imshow(correlation_matrix, color_continuous_scale='Viridis',
                                    title='Correlation heatmap',
                                    labels=dict(color='Correlation'))

external_stylesheets = ['https://cdn.jsdelivr.net/npm/bootswatch@4.5.2/dist/litera/bootstrap.min.css']

fig_population_beforeww2.update_layout(title_x=0.5)
fig_population_afterww2.update_layout(title_x=0.5)
fig_population_over_time.update_layout(title_x=0.5)
fig_pie_comparison.update_layout(title_x=0.5)
fig_correlation_heatmap.update_layout(title_x=0.5)


app = dash.Dash(__name__, external_stylesheets=external_stylesheets)

app.layout = html.Div(children=[
    html.H1(children='Russia Population Visualization', style={'font-weight': 'bold','font-family': 'Montserrat, Arial, sans-serif', 'text-align': 'center'}),

    html.H1(children='Column Visualization', style={'font-family': 'Montserrat, Arial, sans-serif', 'text-align': 'left'}),

    dcc.Dropdown(
        id='column-dropdown',
        options=[
            {'label': col, 'value': col} for col in df_pop.columns
        ],
        multi=True,
        value=['Average population'],
        style={'width': '50%'}
    ),

    dcc.Graph(
        id='column-visualization'
    ),

    html.H1(children='Correlation Heatmap', style={'font-family': 'Montserrat, Arial, sans-serif', 'text-align': 'left'}),
    
    dcc.Graph(
        id='corr-heatmap',
        figure=fig_correlation_heatmap
    ),

    html.Div(children=[
        html.H1(children='Population Before World War 2', style={'font-family': 'Montserrat, Arial, sans-serif', 'text-align': 'left'}),
        dcc.Graph(
            id='population-before-ww2',
            figure=fig_population_beforeww2
        ),
        html.H1(children='Population After World War 2', style={'font-family': 'Montserrat, Arial, sans-serif', 'text-align': 'left'}),
        dcc.Graph(
            id='population-after-ww2',
            figure=fig_population_afterww2
        ),
    ]),

    html.H1(children='Total Population', style={'font-family': 'Montserrat, Arial, sans-serif', 'text-align': 'left'}),

    dcc.Graph(
        id='total-population-chart',
        figure=fig_population_over_time
    ),

    html.H1(children='Population Before & After World War 2', style={'font-family': 'Montserrat, Arial, sans-serif', 'text-align': 'left'}),

    dcc.Graph(
        id='population-pie-comparison',
        figure=fig_pie_comparison
    )
])
```

```python
@app.callback(
    Output('column-visualization', 'figure'),
    [Input('column-dropdown', 'value')]
)

def update_graph(selected_columns):
    if not selected_columns:
        selected_columns = ['Average population']

    if len(df_pop['Year']) != len(df_pop[selected_columns[0]]):
        raise ValueError("Selected columns must have the same length as 'Year'.")

    fig = px.line(df_pop, x='Year', y=selected_columns, title='Column Visualization')
    return fig
```
